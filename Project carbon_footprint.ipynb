```python
import numpy as np

np.mean(var['transport_km'])
np.median(var['Electricity_kwh'])
np.std(var['Plastic_waste_kg'])
np.corrcoef(var['transport_km'], var['Electricity_kwh'])  # Correlation
```

```python
#Top 5 highest footprint days:
var['Total_footprint'] = var['transport_km'] + var['Electricity_kwh'] + var['Plastic_waste_kg']
var.sort_values('Total_footprint', ascending=False).head()
```

```python
1. ðŸ“… Monthly Total Footprint Trend (Line Chart)
Visualize how footprint changes month by month.
```

```python
import matplotlib.pyplot as plt
monthly = var.groupby(var['DATE'].dt.to_period('M')).sum(numeric_only=True)
monthly['Total_footprint'] = monthly['transport_km'] + monthly['Electricity_kwh'] + monthly['Plastic_waste_kg']

monthly.index = monthly.index.to_timestamp()
plt.plot(monthly.index, monthly['Total_footprint'], marker='o')
plt.title("Monthly Total Carbon Footprint")
plt.xlabel("Month")
plt.ylabel("Total Footprint")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
var['DATE'] = pd.to_datetime(var['DATE'], format='%d-%m-%Y')
```

2. User-wise Footprint Comparison (Bar Chart)
Compare users by their average or total footprint.

```python
user_avg = var.groupby('User_Id')[['transport_km', 'Electricity_kwh', 'Plastic_waste_kg']].mean()
user_avg['Total'] = user_avg.sum(axis=1)
user_avg.sort_values('Total', ascending=False, inplace=True)

user_avg['Total'].head(10).plot(kind='bar', color='teal')
plt.title("Top 10 Users by Avg Footprint")
plt.ylabel("Average Footprint")
plt.xlabel("User_ID")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
3. Pie Chart of Food Type Contribution
How much each food type contributes to the total footprint?
```

```python
food_footprint = var.groupby('Food_type')[['transport_km', 'Electricity_kwh', 'Plastic_waste_kg']].sum()
food_footprint['Total'] = food_footprint.sum(axis=1)

plt.pie(food_footprint['Total'], labels=food_footprint.index, autopct='%1.1f%%', startangle=140)
plt.title("Footprint Share by Food Type")
plt.axis('equal')
plt.show()
```

```python
4.Scatter Plot: Transport vs Electricity
Check if high transport users also consume more electricity.
```

```python
plt.scatter(var['transport_km'], var['Electricity_kwh'], alpha=0.4)
plt.title("Transport vs Electricity Usage")
plt.xlabel("Transport (km)")
plt.ylabel("Electricity (kWh)")
plt.grid(True)
plt.show()
```

```python
6.Multiple Histograms (Subplots)
Check distributions for each component side by side.
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].hist(var['transport_km'], bins=20, color='orange')
axs[0].set_title("Transport (km)")

axs[1].hist(var['Electricity_kwh'], bins=20, color='green')
axs[1].set_title("Electricity (kWh)")

axs[2].hist(var['Plastic_waste_kg'], bins=20, color='blue')
axs[2].set_title("Plastic Waste (kg)")

for ax in axs:
    ax.set_ylabel("Frequency")
plt.tight_layout()
plt.show()
```
